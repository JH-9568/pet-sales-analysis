# 주제 : 경기도 내의 반려동물 업체들이 분포한 위치의 유동인구에 따른 폐업률의 상관관계

# 2022105494 소프트웨어융합학과 정진형 

 (1 )주제 선정 이유 : 반려동물들을 파는 판매 업체들이 유동인구가 많거나 적은 곳에 있을 때 판매업체들의 폐업률이 높고 낮은지를 분석하고 , 유동인구와 폐업률의 상관관계를 분석하여 스타트업이 인기를 얻고있는 요즘 시대에 스타트업을 시작하기에 적합한 위치를 알아보기 위하여 이 주제를 선정하였다.

(2) 가설 정의 : 가설 - 유동인구가 많은 지역이나 번화가 근처에 위치한 반려동물판매업체들의 폐업률이 비교적 유동인구가 적은 지역에 있는 업체들보다 폐업률이 낮을 것이다. 

(3) 인터넷을 통한 데이터 획득 : 정부 3.0 공공데이터 사이트를 활용하여 '경기도 내 반려동물 판매업체 현황' 에 대한 데이터를 수집하였다. 이 데이터 속에는 여러 경기도 반려동물 판매업체의 위치(시군명,도로명우편번호,소재지도로명주소,소재지지명주소) 를 포함하고 있으며 각 판매업체의 영업상태 ( 만약 폐업하였을 경우 폐업일자를 보여줌 ) 또한 포함하고 있다. 그리고 정부 3.0 공공데이터 사이트를 이용한 '경기도_시군별 유동인구 데이터' 를 활용하였다. 이 데이터에는 경기도 내의 시군별 유동인구를 보여준다. 

```python
import pandas as pd
df=pd.read_json("반려동물판매업체현황.json")
print(df) # "반려동물판매업체현황 데이터 "
```
